```python
d = 5
centers, delta = ATS.box_centers(X_dgm1_train, d, padding=0.05) 

# Here we are computing the features for each tent function 
t0 = time.time()

X_train_features_1_tent = ATS.get_all_features_boxes(X_dgm1_train, centers, delta)

X_test_features_1_tent = ATS.get_all_features_boxes(X_dgm1_test, centers, delta)

t1 = time.time()
print('Computing features from H_1 took (seconds):{}'.format(t1-t0))
```

### Adaptive features GMM

```python
print('Begin GMM...')
t0 = time.time()
X_train_temp = np.vstack(X_dgm1_train)

gmm_f_train=[]
for i in range(len(X_dgm1_train)):
    gmm_f_train.append(y_train[i]*np.ones(len(X_dgm1_train[i])))
gmm_f_train = np.concatenate(gmm_f_train)

gmm = mixture.BayesianGaussianMixture(n_components=d*d, covariance_type='full', max_iter=int(10e4)).fit(X_train_temp, gmm_f_train)

ellipses = []
for i in range(len(gmm.means_)):
    L, v = np.linalg.eig(gmm.covariances_[i])
    temp = {'mean':gmm.means_[i], 'std':np.sqrt(L), 'rotation':v.transpose(), 'radius':max(np.sqrt(L)), 'entropy':gmm.weights_[i]}
    temp['std'] = 3*temp['std']
    ellipses.append(temp)
t1 = time.time()
print('Finish GMM. Time: {}'.format(t1-t0))

# -----------------------------------------------------------------------------
# ------------------------------ GMM features ---------------------------------
# -----------------------------------------------------------------------------

t0 = time.time()

X_train_features_1_gmm = ATS.get_all_features(X_dgm1_train, ellipses, ATS.f_ellipse)

X_test_features_1_gmm = ATS.get_all_features(X_dgm1_test, ellipses, ATS.f_ellipse)

t1 = time.time()
print('Computing features from H_1 took (seconds):{}'.format(t1-t0))
```

### HDB Scan

```python
# -----------------------------------------------------------------------------
# ------------------------------ HDBSCAN --------------------------------------
# -----------------------------------------------------------------------------

print('Begin HDBSCAN...')
t0 = time.time()
X_train_temp = np.vstack(X_dgm1_train)

clusterer = hdbscan.HDBSCAN()

clusterer.fit(X_train_temp)

num_clusters = clusterer.labels_.max()

ellipses = []
for i in range(num_clusters):
    cluster_i = X_train_temp[clusterer.labels_ == i]

    en = np.mean(clusterer.probabilities_[clusterer.labels_ == i])

    mean = np.mean(cluster_i, axis=0)
    cov_matrix = np.cov(cluster_i.transpose())

    L,v = np.linalg.eig(cov_matrix)

    temp = {'mean':mean, 'std':np.sqrt(L), 'rotation':v.transpose(), 'radius':max(np.sqrt(L)), 'entropy':en}
    temp['std'] = 2*temp['std']
    ellipses.append(temp)

t1 = time.time()
print('Finish HDBSCAN. Time: {}'.format(t1-t0))

# -----------------------------------------------------------------------------
# ------------------------------ HDBSCAN features -----------------------------
# -----------------------------------------------------------------------------

t0 = time.time()

X_train_features_1_hdbscan = ATS.get_all_features(X_dgm1_train, ellipses, ATS.f_ellipse)

X_test_features_1_hdbscan = ATS.get_all_features(X_dgm1_test, ellipses, ATS.f_ellipse)

t1 = time.time()
print('Computing features from H_1 took (seconds):{}'.format(t1-t0))
```

### CDER

```python
# -----------------------------------------------------------------------------
# ------------------------------ CDER -----------------------------------------
# -----------------------------------------------------------------------------

#colors = ['red', 'yellow', 'magenta', 'green', 'blue', 'black']

y_train_cder = y_train.copy()
n = len(label_unique)
print(n)

print('Begin CDER...')
t0 = time.time()

pc_train = multidim.PointCloud.from_multisample_multilabel(X_dgm1_train, y_train_cder)
ct_train = CoverTree(pc_train)

cder = CDER(parsimonious=True)

cder.fit(ct_train)

cder_result = cder.gaussians

t1 = time.time()
print('Finish CDER. Time: {}'.format(t1-t0))

ellipses = []
for c in cder_result:
    temp = {key:c[key] for key in ['mean', 'std', 'rotation', 'radius', 'entropy']}
    temp['std'] = 3*temp['std']
    ellipses.append(temp)

# -----------------------------------------------------------------------------
# ------------------------------ CDER features --------------------------------
# -----------------------------------------------------------------------------
t0 = time.time()

X_train_features_1_cder = ATS.get_all_features(X_dgm1_train, ellipses, ATS.f_ellipse)

X_test_features_1_cder = ATS.get_all_features(X_dgm1_test, ellipses, ATS.f_ellipse)

t1 = time.time()
print('Computing features from H_1 took (seconds):{}'.format(t1-t0))
```

### Kernel

```python
n_train = len(X_dgm1_train)
n_test = len(X_dgm1_test)
X_train_features_1_kernel = np.zeros((n_train, n_train))
X_test_features_1_kernel = np.zeros((n_test, n_train))
s = 1

start = time.time()
for i in range(0,n_train):
    print(i)
    for j in range(0,i):
        dgm0 = X_dgm1_train[i]
        dgm1 = X_dgm1_train[j]
        hk = heat(dgm0, dgm1, sigma = s)        
        X_train_features_1_kernel[i,j] = hk
        X_train_features_1_kernel[j,i] = hk
    for j in range(0,n_test):
        dgm0 = X_dgm1_train[i]
        dgm1 = X_dgm1_test[j]
        hk = heat(dgm0, dgm1, sigma = s)        
        X_test_features_1_kernel[j,i] = hk
        
time.time()-start
```
